```python
def plot_cosine_similarity(cosine_similarity, ax, model_name: str) -> None:
    """
    Plots a spectrogram of the cosine similarity matrix on a given axis.
    """
    ax.set_title(model_name, fontsize=10)
    cax = ax.imshow(cosine_similarity, aspect='auto', cmap='hot', interpolation='nearest')
    ax.set_xlabel('Sentence Index')
    ax.set_ylabel('Sentence Index')
    colorbar(cax, ax=ax, label='Cosine Similarity')
    print(f'Model: {model_name}, expectation: {mean(cosine_similarity):.4f}, variance: {var(cosine_similarity):.4f}')
```

$
\text{Cosine Similarity} = \frac{|A| \times |B|}{A \cdot B}
$

```python
def jaccard_similarity(doc1, doc2):
    set1 = set(doc1.split())
    set2 = set(doc2.split())
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union

jaccard_similarities = zeros((len(documents), len(documents)))
for i in range(len(documents)):
    for j in range(len(documents)):
        jaccard_similarities[i][j] = jaccard_similarity(documents[i], documents[j])

fig, ax = subplots(1, 1, figsize=(8, 6))
cax = ax.imshow(jaccard_similarities, aspect='auto', cmap='hot', interpolation='nearest')
ax.set_title('Jaccard Similarity Matrix', fontsize=10)
ax.set_xlabel('Sentence Index')
ax.set_ylabel('Sentence Index')
colorbar(cax, ax=ax, label='Jaccard Similarity')
show()
```
